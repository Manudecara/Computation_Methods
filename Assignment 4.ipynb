1.) Proof that the following statements are equivalent:

1. $z$ is an eigenvalue of $A+\delta A$ for some $\delta A$ with $\|\delta A\|_2\leq\epsilon$.
2. There exists a vector $u\in\mathbb{C}^m$ with $\|(A-zI)u\|_2\leq \epsilon$ and $\|u\|_2=1$.
3. $\sigma_n(zI-A)\leq\epsilon$, where $\sigma_n$ is the smallest singular value of $zI-A$.
4. $\|(zI-A)^{-1}\|_2\geq\epsilon^{-1}$.

**Solution** 

$\bf{i)\rightarrow ii)}$: Assuming that statement $\bf{(i)}$ holds: 

If $z$ is an eigenvalue of $A+\delta A$ where $\|\delta A\|_2\leq\epsilon$ then $\exists u\in\mathbb{C}^n$ a right eigenvector with $\|u\|_2=1$ such that: 
$$ (A+\delta A)u = zu$$ 
Rearranging the above equation leads to: 
$$ (A-zI)u=-\delta Au$$ 
Taking Euclidean norm of both sides of the equation leads to: 
$$ \|(A-zI)u\|_2 = \|\delta Au\|_2 \leq \|\delta A\|_2\|u\|_2 = \|\delta A\|_2 \leq \epsilon $$
$$ \|(A-zI)u\|_2 \leq \epsilon \\ $$ 

###### ==================================================================================================================

$\bf{ii)\rightarrow iii)}$: Let $(zI-A)\in\mathbb{C}^{m\times m}$

We can calculate a $\underline{Full\ SVD}$ of matrix $(zI-A)$ such that $(zI-A) = U\Sigma V^H$ where $U,V\in\mathbb{C}^{m\times m}$ are square unitary matrices and $\Sigma = \begin{bmatrix}\tilde{\Sigma} & 0\\ 0 & 0\end{bmatrix}\ \text{where} \ \tilde{\Sigma} = diag(\sigma_1,\dots,\sigma_n) $ is a diagonal matrix where $\sigma_1\geq \sigma_2\geq \dots \geq \sigma_n > 0\quad (n\leq m)$ 

Let us now consider: 
$$\|(zI-A)u\|_2 = \|U\Sigma V^Hu\|_2$$
$$= \|\Sigma V^Hu\|_2$$
But: 
$$\|\Sigma V^Hu\|_2^2 = \sum_{j=1}^m(\sigma_jv_j^Hu)^H(\sigma_jv_j^Hu) = \sigma_n^2\sum_{j=1}^m\left(\frac{\sigma_j}{\sigma_n}\right)^2\left(v_j^Hu\right)^H\left(v_j^Hu\right)$$

$\sigma_n$ represents the smallest singular value of $(zI-a) \implies \frac{\sigma_j}{\sigma_n} \geq 1 \ \text{for} \  j=1,\dots, n \ \text{where} \ \sigma_n \neq 0 $ as per the definition of SVD. Therefore: 

$$ \|\Sigma V^Hu\|_2^2 \geq \sigma_n^2\sum_{j=1}^m(v_j^Hu)^H(v_j^Hu)$$
$$ \Rightarrow \|\Sigma V^Hu\|_2 \geq \sigma_n\|V^Hu\|_2 = \sigma_n$$ 
since $\|V^H\|_2 =1 \ , \ \|u\|_2 = 1.$

$\implies \|\Sigma V^Hu\|_2 = \|(zI-A)u\|_2 \geq \sigma_n$ 

But:
$$ \sigma_n \leq \|(zI-A)u\|_2 \leq \epsilon $$ 
$$\Rightarrow \sigma_n(zI-A)\leq\epsilon$$
###### ==================================================================================================================
$\bf{iii)\rightarrow iv)}$: $z$ is **not** an eigenvalue of $A$, thus $det(zI-A)\neq 0$ and $[zI-A]^{-1}$ exists. 

From $\bf{(iii)}$: $\|zI-A\|_2 = \sigma_n(zI-A) \implies \|(zI-A)^{-1}\|_2 = [\sigma_n(zI-A)]^{-1} \geq \epsilon^{-1}$  

$$\|(zI-A)^{-1}\|_2 \geq \epsilon^{-1} $$
###### ==================================================================================================================
$\bf{iv)\rightarrow i)}$: By the definition of **Euclidean Norm**
$$\displaystyle\|A\|_{2}:=\max_{x\in\mathbb{R}^n\backslash\{0\}} \frac{\|Ax\|_2}{\|x\|_2}$$
We can assume that $\|x\|_2=1 \implies $:
$$\displaystyle\|A\|_{2}:=\max_{x\in\mathbb{R}^n\backslash\{0\}} \|Ax\|_2$$

From $\bf{(iv)}$ we know that: $$\|(zI-A)^{-1}\|_2 \geq \epsilon^{-1} \iff \max_{x\in\mathbb{R}^n\backslash\{0\}} \|(zI-A)^{-1}x\|_2 \geq \epsilon^{-1} \ \text{provided}\ \|x\|_2=1 $$

Let's choose $\hat{\epsilon}^{-1} \geq \epsilon^{-1}$ such that $\|(zI-A)^{-1}\|_2 = \hat{\epsilon}^{-1}\geq \epsilon^{-1} \implies \hat{\epsilon}\leq \epsilon$

Let us define $v\in\mathbb{C}^m$ such that: 
$$ v = (zI-A)^{-1}x$$
But $\|(zI-A)^{-1}x\|_2 = \hat{\epsilon}^{-1} \implies \|v\|_2=\hat{\epsilon}^{-1} \implies \|v\|_2\hat{\epsilon}=1$

So if we choose $u=\hat{\epsilon}v \implies \|u\|_2 = 1$
Thus we can say:
$$u=\hat{\epsilon}v = \hat{\epsilon}(zI-A)^{-1}x$$
$$u=\hat{\epsilon}(zI-A)^{-1}x$$

$$\Rightarrow \ (zI-A)u= \hat{\epsilon}x$$
$$Au+\hat{\epsilon}x = zu$$
$$(A+\underbrace{\hat{\epsilon}xu^H}_{\delta A})u = zu$$ 

If we choose $\delta A = \hat{\epsilon}xu^H$ then $\|\delta A\|_2 = \hat{\epsilon} \leq \epsilon \implies $
$$ (A+\delta A)u = zu$$

Therefore $z$ is a right eigenvalue of $A+\delta A$ for some $\delta A$ with $\|\delta A\|_2\leq\epsilon \quad \square$

                                
                                
                              ***Thus we have shown that the following statements are equivalent***

2.) Let $\lambda$ be a simple eigenvalue of a diagonalizable matrix $A$ with right and left eigenvectors $x$ and $y$, and let $\lambda+\delta\lambda$ be the corresponding eigenvalue of the matrix $A+\delta A$. Show that to first order

$$
\delta\lambda = \frac{y^H\delta Ax}{y^Hx}.
$$

**Solution**

For a diagonalizable matrix $A$:

$\underline{\text{Right Eigenvectors}}: \quad \bf{Ax=\lambda x} $

$\underline{\text{Left Eigenvectors}}: \quad \bf{y^HA=\lambda y^H}$ 

We would like to understand how $\lambda$ changes under small perturbations to $A$. If we consider the equation $ (A+\delta A)x = \lambda x$ and parametrize the equation with a parameter $t$. We get: 
$$ (A+\delta At)x(t)=\lambda (t) x(t) \quad (\ast)$$

If we formally differentiate the eigenvalue problem $(\ast)$ with respect to $t$ then we have: 

$$ (\delta A)x(t) + A(\delta x(t)) = (\delta \lambda(t))x(t) + \lambda(t)(\delta x(t))$$
**Note**: here I ignored the second order term $\delta A\delta x(t)t$. 

Just for simpler notation, let $\lambda (t) = \lambda \ , \ x(t)=x $

If we multiply this expression by $y^H$ on the left, then we have:

$$  y^H(\delta A)x + y^HA(\delta x) = y^H(\delta \lambda)x + y^H\lambda(\delta x)$$
**Note**: $y^HA=\lambda y^H\implies y^HA(\delta x)=\lambda y^H(\delta x)$ so that we have: 

$$y^H(\delta A)x = (\delta \lambda)y^Hx $$
which we rearrange to get

$$\bf{\delta\lambda = \frac{y^H\delta Ax}{y^Hx}} \quad \square$$

3.) Consider the integral $I(f) := \int_a^bf(x)dx$. Assume that we want to approximate $I(f)$ by a Newton-Cotes quadrature rule of the form $I_2 :=\omega_0f(a) + \omega_1f((a+b)/2) + \omega_2f(b)$. Derive the corresponding quadrature weights $\omega_0$, $\omega_1$ and $\omega_2$.

**Solution**

$$\int_a^bf(x)\ dx\approx \int_a^bp_n(x)\ dx  =\int_a^b\sum_{k=0}^nL_k(x)f(x_k)\ dx =\sum_{k=0}^nf(x_k)\int_a^bL_k(x)\ dx  $$
$$ \Rightarrow \int_a^bf(x)\ dx \approx \sum_{k=0}^nf(x_k)\underbrace{\int_a^b \prod_{j=0,j\neq k}^n \frac{(x-x_j)}{(x_k-x_j)}\ dx}_{\omega_k} = \sum_{k=0}^nf(x_k)\omega_k $$

Let us define: 
$$x_i = a+ih \quad \quad i=0\dots n$$
where $h = \frac{b-a}{n}$ 

Since $\bf{n=2} \implies h= \frac{b-a}{2}$ resulting in:
$$x_0 = a \quad \quad x_1 = \frac{b+a}{2} \quad \quad x_2 = b$$
being the equipspaced points for Simpsons Rule. 

Let us consider the derivation of $\omega_0 \ ,\ \omega_1 \ , \ \omega_2$:

$$\omega_0  =  \int_a^b\frac{(x-x_1)(x-x_2)}{(x_0-x_1)(x_0-x_2)}\ dx $$
$$ = \frac{1}{(a-b)^2} \int_a^b (2x-(a+b))(x-b) \ dx $$

Let $x= \frac{a+b}{2}+u$ 

$$\Rightarrow \omega_0 \ = \frac{1}{(a-b)^2} \int_{-h}^h 2u(u+\frac{a+b}{2}-b) \ du$$
$$ = \frac{1}{(a-b)^2} \int_{-h}^h 2u(u-h) \ du $$
$$ = \frac{2}{(a-b)^2}\left[\frac{1}{3}u^3 - \frac{1}{2}u^2h\right]^{h}_{-h}$$
$$ = \frac{2}{(a-b)^2}\left[\frac{2}{3}h^3\right] $$
$$ = \frac{4}{3}.\frac{1}{(a-b)^2}.\left(\frac{(b-a)}{2}\right)^3$$
$$ =\frac{b-a}{6}$$

We now use the symmetric properties of Newton-Cotes rules to derive that $\bf{\omega_2 = \omega_0 = \frac{b-a}{6}}$. We also know that sum of the weights $\omega_k$ is always $(b - a) \implies \bf{\omega_1 = \frac{4(b-a)}{6}}$

Since $\quad \int_a^bf(x)dx \approx \omega_0f(a) + \omega_1f(\frac{a+b}{2}) + \omega_2f(b):$

$$ \bf{\int_a^bf(x)dx \approx \frac{b-a}{6}\left[\ f(a)+4\ f\left(\frac{a+b}{2}\right)+f(b)\right]} \quad \Box$$

This is the derivation of ***Simpson's 1/3 Rule*** .

## Programming Part

In this programming exercise we will investigate the numerical properties of Lagrange interpolation.
To make things easy you can use the function ``scipy.interpolate.BarycentricInterpolator`` to create the Lagrange
interpolation polynomial. Please see the `scipy` help entries for its use.

We consider the following functions in the interval $[-1,1]$.

* $f(x)=|x|$
* $f(x)=\frac{1}{1+10x^2}$

Write a script which for each function evaluates the maximum error between the Lagrange interpolation polynomial $p_n$ in $[-1,1]$ and the function $f$, and plot the corresponding convergence graph of the error against the polynomial degree $n$. You may want to use fully logarithmic or semilogarithmic plots for this. In order to evaluate the maximum error the most simple thing is to evaluate the function $f$ and the polynomial $p_n$ at a few hundred equispaced points in $[-1,1]$ and use the maximum difference as approximation to the error over the whole interval.

As interpolation points you should try equispaced points and Chebychev points. You find the definition of Chebychev points at https://en.wikipedia.org/wiki/Chebyshev_nodes. Describe your observations with respect to convergence and Runge phenomenon.

```python
                 '''Computing the Interpolation Error when choosing Equispaced and Chebychev Points'''

import numpy as np 
import matplotlib.pyplot as plt 
import scipy.interpolate 

def interpolation_error(f): 
    deg = range(1,100)
    xeval = np.linspace(-1,1,400)
    error = []    #Create an empty array to store maximum error of equispaced points
    cheberror = [] #Create an array to store maximum error of chebychev points
    f_exact = f(xeval)   #Calculating the Exact function 
    for n in deg: 
        x = np.linspace(-1,1,n)  #create n equispaced points 
        cheb = [np.cos((2.*k-1)/(2.*n)*np.pi) for k in range(1,n+1)] #creating n chebychev points
        chebf = [f(np.cos((2.*k-1)/(2.*n)*np.pi)) for k in range(1,n+1)] #performing the function on these points
        poly1 = scipy.interpolate.BarycentricInterpolator(x,f(x))   #Lagrange Interpolation Polynomials for equipspaced points
        poly2 = scipy.interpolate.BarycentricInterpolator(cheb,chebf)   #Lagrange Interpolation Polynomial for Chebychev points
        error.append(np.max(abs(f_exact-poly1(xeval))))   #maximum error for given degree of equispaced points
        cheberror.append(np.max(abs(f_exact-poly2(xeval))))  #maximum error for given degree of chebychev points
    return deg, error, cheberror

func1 = lambda x: abs(x)
func2 = lambda x: 1/(1+10*x**2)

''' Graph for Equispaced Points'''
    
plt.semilogy(interpolation_error(func1)[0],interpolation_error(func1)[1],'k-x',label = '$f(x) = |x|$');
plt.semilogy(interpolation_error(func2)[0],interpolation_error(func2)[1],'r-x',label = '$f(x) = 1/(1+10x^2)$');
plt.title('Maximum Error in Equispaced points')
plt.legend()
plt.xlabel('Degree $n$ of polynomial $p_n$');
plt.ylabel('Maximum Error $e_n$')
plt.gcf().set_size_inches(10.5, 8)
plt.show();
```

**Equipspaced Points**

It is clearly visible that when we picked equispaced points, the maximum absoluted error $e_n = \max_{-1\leq x \leq 1} |f(x) - p_n(x)|$ diverges exponenetially with higher degree polynomials $p_n$. This is due to Runge's Phenomena where Carl David Tolmé Runge (1901) showed that going to higher degrees does not always improve accuracy -  oscillation at the edges of an interval when using equipspaced points can lead to huge errors. 


```python
                                         '''Graph for Chebychev points'''

plt.semilogy(interpolation_error(func1)[0],interpolation_error(func1)[2],'b-x',label = '$f(x) = |x|$');
plt.title('Maximum Error in Chebychev points for f(x) = |x|')
plt.legend()
plt.xlabel('Degree $n$ of polynomial $p_n$');
plt.ylabel('Maximum Error $e_n$')
plt.gcf().set_size_inches(10.5, 8)
plt.show()
plt.semilogy(interpolation_error(func2)[0],interpolation_error(func2)[2],'r-x',label = '$f(x) = 1/(1+10x^2)$');
plt.title('Maximum Error in Chebychev points for $f(x) = 1/(1+10x^2)$')
plt.legend()
plt.xlabel('Degree $n$ of polynomial $p_n$');
plt.ylabel('Maximum Error $e_n$')
plt.gcf().set_size_inches(10.5, 8)
```

**Chebychev Points**

Maximum absolute error $e_n$ converges nicely when using Chebychev points. This is because chebychev points are distributed more densely towards the edges of the interval, minimising the oscillation. $g(x)=|x|$ converges at a rate of $\bf{\frac{1}{N^{\alpha}}} $ with $\alpha > 1$ where $e_n \approx CN^{-\alpha} \implies log(e_n)\approx \frac{C}{N^{\alpha}}$ when using ``matplotlib.pyplot.semilogy`` on python.$\ f(x)=\frac{1}{1+10x^2}$ converges **exponentially** where $e_n\approx CR^{-N} \implies log(e_n) \approx log(C)-Nlog(R)$ (straight line in my plot) where$\ R>1  \ , \ C>0$.  We can approximate $f(x)=\frac{1}{1+10x^2}$ with a higher accuracy over $g(x)=|x|$ because $e_n$ for $f(x)$ converges at a faster rate for a given degree $n$.

```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.interpolate

def absolute(x):
    return abs(x)

def func(x):
    return 1/(1+10*x**2)

def chebychev(n,f):
    temp =[]
    tempf = []
    for k in np.arange(1,n+1):
        temp.append(np.cos((2.*k-1)/(2.*n)*np.pi))
        tempf.append(f(np.cos((2.*k-1)/(2.*n)*np.pi)))
    return temp,tempf 

eval_x = np.linspace(-1,1,100)
y1_exact = absolute(eval_x)
y2_exact = func(eval_x)
degree = np.arange(1,101)
equi_error1 = []
equi_error2 = []
cheb_error1 =[]
cheb_error2 =[]

for n in degree: 
    x = np.linspace(-1,1,n)
    approx1 = scipy.interpolate.BarycentricInterpolator(x,absolute(x))
    approx2 = scipy.interpolate.BarycentricInterpolator(x,func(x))
    equi_error1.append(np.max(abs(approx1(eval_x)-y1_exact)))
    equi_error2.append(np.max(abs(approx2(eval_x)-y2_exact)))
    cheb1 = chebychev(n,absolute)[0]
    chebf1 = chebychev(n,absolute)[1]
    cheb2 = chebychev(n,func)[0]
    chebf2 = chebychev(n,func)[1]
    approx_cheb1 = scipy.interpolate.BarycentricInterpolator(cheb1,chebf1)
    approx_cheb2 = scipy.interpolate.BarycentricInterpolator(cheb2,chebf2)
    cheb_error1.append(np.max(abs(approx_cheb1(eval_x)-y1_exact)))
    cheb_error2.append(np.max(abs(approx_cheb2(eval_x)-y2_exact)))
print()
print('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
print()
print('**Maximum error for equispaced points**')
plt.semilogy(degree,equi_error1,'k-x',label = 'f(x)=|x|');
plt.legend()
plt.show()
plt.semilogy(degree,equi_error2,'r-x',label = 'f(x)=1/(1+10x^2)');
plt.legend()
plt.show()
print('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
print()
print('**Maximum error for chebychev points**')
plt.semilogy(degree,cheb_error1,'k-x',label = 'f(x)=|x|');
plt.legend()
plt.show()
plt.semilogy(degree,cheb_error2,'r-x',label = 'f(x)=1/(1+10x^2)');
plt.legend()
plt.show()
```
