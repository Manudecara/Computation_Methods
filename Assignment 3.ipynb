**1a) Given the least-squares problem 
$$
\|A\hat{x}-b\|_2 = \min_{x\in
\mathbb{R}^{n}}\|Ax-b\|_2
$$
with $A\in\mathbb{R}^{m\times n}$, $m \geq n$ and $\text{rank}(A) = n$.**

**Show that the solution $\hat{x}$ is given by the solution of the linear system of equations
$$
\begin{bmatrix}
I & A\\
A^T & 0
\end{bmatrix}\begin{bmatrix}
y\\ x \end{bmatrix}=\begin{bmatrix}b \\ 0\end{bmatrix}.
$$
b) What is the interpretation of $y$?**



**Solution**

**a)** We know that Least Square Problem must satisfy the Normal Equation: 
$$A^TAx=A^Tb \quad \implies \quad A^T(b-Ax)=0$$

Let $y\in\mathbb{R}^{m}$ such that $y=b-Ax$. Thus we need to satisfy two conditions: 
$$y+Ax = b \quad \text{and} \quad A^Ty=0$$
which can be expressed as a linear systems of equations: 

$$
\begin{bmatrix}
I & A\\
A^T & 0
\end{bmatrix}\begin{bmatrix}
y\\ x \end{bmatrix}=\begin{bmatrix}b \\ 0\end{bmatrix}
\quad \quad \quad \square$$

**b)** $y=b-Ax$ represents the component of vector $b$ that is $\bf{orthogonal}$ to the $\textbf{span\{A\}}.$ In other words, $\|y\|_2$ represents the $\textbf{smallest error}$ of the Least Squares Problem. 





**2.) Let $A=QR$ be the reduced QR factorization of the matrix $A$ in the above least-squares problem. Show that the solution of the least-squares problem can be written as $x=R^{-1}Q^Tb$. Then show that
$$\|\hat{x}\|_2\leq\sigma_n^{-1}\|b\|_2,$$ where $\sigma_n$ is the smallest singular value of $A$.**

**Solution**

**a)** The $\textbf{Reduced QR Factorization}$ of $A\in\mathbb{R}^{m\times n}$ is $A=QR$ where $Q\in\mathbb{R}^{m\times n}$ is an Orthogonal matrix ($Q^TQ=I$) and $R\in\mathbb{R}^{n\times n}$ is an upper triangular square matrix. 

Best way to solve Least Sqaures Problem is to solve the Normal Equation: 
$$A^TAx=A^Tb$$   
$$ (QR)^TQRx=(QR)^Tb $$
$$R^TRx=R^TQ^Tb$$

Since $A$ has a $full\ rank$, all the terms in the leading diagonal of $R$ i.e. all $r_{ii} \neq 0$ where $1\leq i\leq n.$ Thus $\implies R^{-1} \text{and} \ (R^{T})^{-1}$ exists. So we can simplify the steps down further. 
$$Rx=Q^Tb$$
$$x=R^{-1}Q^Tb \quad \square$$

**b)** We can decompose $A\in\mathbb{R}^{m\times n}$ into a $\bf{full}$ SVD where $U\in\mathbb{R}^{m\times m}$, $\Sigma\in\mathbb{R}^{m\times n}$ and $V\in\mathbb{R}^{n\times n}$ such that $A=U\Sigma V^T$. 
Both $U$ and $V$ are $\textbf{orthogonal square matrices}$ : 
$$U^TU = UU^T =I= V^TV = VV^T$$ 

$\Sigma = \begin{bmatrix}\tilde{\Sigma}\\ 0\end{bmatrix} \quad \text{where} \quad \tilde{\Sigma} = diag(\sigma_1,\dots,\sigma_n)$

Considering the Normal Equation: 
$$A^TAx=A^Tb$$ 
$$(U\Sigma V^T)^T(U\Sigma V^T)x = (U\Sigma V^T)^Tb$$ 
$$V\Sigma^TU^TU\Sigma V^Tx = V\Sigma^TU^Tb$$
$$V\Sigma^T\Sigma V^Tx = V\Sigma^T U^T b$$   
$$\Sigma^T\Sigma V^Tx = \Sigma^T U^T b$$ 

$A$ has a $full \ rank:$ $n$ i.e. all the singular values $\sigma_i \neq 0$ and $\Sigma^T\Sigma\in\mathbb{R}^{n\times n}$ is a square matrix such that $det(\Sigma^T\Sigma) \neq 0$. Hence we've shown that $\Sigma^T\Sigma$ is an invertible matrix. Clearly $V^T$ is also invertible as $V$ is a square matrix both in full and reduced SVD:

$$ x = (\Sigma^T\Sigma V^T)^{-1}\Sigma^T U^T b$$
$$ x = (V^T)^{-1}(\Sigma^T\Sigma )^{-1}\Sigma^TU^Tb$$
$$ x = V(\Sigma^T\Sigma )^{-1}\Sigma^TU^Tb \quad (\ast)$$
Let $\bf{K = (\Sigma^T\Sigma )^{-1}\Sigma^T}$. Clearly 
$$(\Sigma^T\Sigma)^{-1}\in\mathbb{R}^{n\times n} = diag(\sigma_1^{-2},\dots,\sigma_n^{-2}) \implies K = \begin{bmatrix}diag(\sigma_1^{-1},\dots,\sigma_n^{-1}) & 0_{(n\times (m-n))}\end{bmatrix}$$
Note that $K^TK$ is a $m\times m$ square matrix such that: 
$$K^TK = \begin{bmatrix}diag(\sigma_1^{-2},\dots,\sigma_n^{-2}) & 0_{(n\times (m-n))}\\
0_{((m-n)\times n)} & 0_{((m-n)\times (m-n))}\end{bmatrix}$$

$$\implies \|K\|_2 = \sqrt{\lambda_{max}(K^TK)} = \sigma_n^{-1}$$
Thus consider $(\ast)$ again: 
$$ x = V(\Sigma^T\Sigma )^{-1}\Sigma^TU^Tb $$
$$ x = VKU^Tb$$ 
Then considering the 2-norms of either side of the equation: 
$$  \|x\|_2 = \|VKU^Tb\|_2$$
$$  \leq \|V\|_2\|K\|_2\|U^T\|_2\|b\|_2$$
$$ = \sigma_n^{-1}\|V\|_2\|U^T\|_2\|b\|_2 $$

**Note** $\quad \|V\|_2 = \sqrt{\lambda_{max}(V^TV)} = \sqrt{\lambda_{max}(I)} = 1 \quad \quad \text{and} \quad \quad \|U^T\|_2 = \sqrt{\lambda_{max}(UU^T)} = \sqrt{\lambda_{max}(I)} = 1$ 

Thus we have shown that: 

$$\bf{\|x\|_2\leq\sigma_n^{-1}\|b\|_2}$$
 where  $\sigma_n$ is the smallest singular value of $A$. $\square$

$\textbf{4.) We consider the modified Least-Squares Problem}$
$$
\min_{x\in\mathbb{R}^n} \|Ax-b\|_2^2+\|L x\|_2^2

$\textbf{for some} $ $L\in\mathbb{R}^{n\times n}$.

* $\textbf{Show that the solution of this least-squares problem is identical to the solution of} $

* $\textbf{Now let }$ $L=\alpha I$ $\textbf{for some}$ $\alpha\neq 0$. $\textbf{Write down an explicit representation of the solution}$ $x$ $\textbf{in terms of }$ $\alpha$ $\textbf{and the singular values and singular vectors of}$ $A$


**a)** The above expression $(\ast)$ can be written as: 
$$\min_{x\in\mathbb{R}^n}\left\|\begin{bmatrix}A\\ L\end{bmatrix}x-\begin{bmatrix}b\\0\end{bmatrix}\right\|_2 \quad \text{where} \quad \tilde{A} = \begin{bmatrix}A\\ L\end{bmatrix} \quad \text{and} \quad \tilde{b} = \begin{bmatrix}b\\ 0\end{bmatrix}$$

Considering the Normal Equation:

$$\tilde{A}^T\tilde{A}x=\tilde{A}^Tb$$
$$={\begin{bmatrix}A\\ L\end{bmatrix}}^T\begin{bmatrix}A\\ L\end{bmatrix}x = {\begin{bmatrix}A\\ L\end{bmatrix}}^T\begin{bmatrix}b\\ 0\end{bmatrix}$$
$$=\left[A^T,L^T\right]\begin{bmatrix}A\\ L\end{bmatrix}x = \left[A^T,L^T\right]\begin{bmatrix}b\\ 0\end{bmatrix}$$
$$=(A^TA+L^TL)x = A^Tb$$


**b)** Now let $A=U\Sigma V^T$ and $L=\alpha I$ for some $\alpha\neq 0$. We also know $U$ and $V$ are Orthogonal Matrices i.e $U^TU=I=V^TV$. Then :


$$(A^TA+L^TL)x = A^Tb$$
$$(V{\Sigma}^TU^TU\Sigma V^T + {\alpha}^2I)x = (U\Sigma V^T)^{T}b$$
$$(V{\Sigma}^T\Sigma V^T + {\alpha}^2I)x = V{\Sigma}^T U^{T}b$$
$$ V(\Sigma^T\Sigma + {\alpha}^2V^TV)V^Tx =  V{\Sigma}^T U^{T}b$$
$$ (\Sigma^T\Sigma + {\alpha}^2I)V^Tx =  {\Sigma}^T U^{T}b$$

$A$ has a $full\ rank:\ n \Rightarrow \forall\sigma_i\neq 0$ where $1\leq i \leq n$ thus the $det(\Sigma^T\Sigma)\neq 0$ and $\Sigma^T\Sigma$ is invertible. $V\in\mathbb{R}^{n\times n}$ is a square orthogonal matrix thus $V^T = V^{-1}$ so we obtain: 

$$x = V{({\Sigma}^T\Sigma  + {\alpha}^2I)}^{-1}{\Sigma}^T U^{T}b$$

$$ = \sum_{j=1}^{n} v_j\frac{\sigma_j u_j^Tb}{\sigma_j^2+\alpha^2} $$

Let $A\in\mathbb{R}^{m \times n}$, $\tilde{U}\in\mathbb{R}^{m\times k}$ and $\tilde{V}\in\mathbb{R}^{n\times k}$ with both $\tilde{U}^T\tilde{U} = I$ and $\tilde{V}^T\tilde{V}=I$. The projection $A_{\tilde{U},\tilde{V}}$ of $A$ onto the spaces spanned by $\tilde{U}$ and $\tilde{V}$ is given by

$$
A_{\tilde{U},\tilde{V}} = \tilde{U} (\tilde{U}^TA\tilde{V}) \tilde{V}^T.
$$

First convince yourself that if $\tilde{U}$ and $\tilde{V}$ consist of the first $k$ left and right singular vectors of $A$ the projection $A_{\tilde{U},\tilde{V}}$ is the low-rank approximation obtained by cutting off the singular vector expansion after the $k$ largest singular values.

Now let $\tilde{U}$ and $\tilde{V}$ be random matrices with $k$ columns.

Take a Scipy picture (such as ``scipy.misc.face(gray=True)``) and project the picture onto your random matrices. Plot the result. You may see that some information from the picture is preserved in this projection onto random matrices.

We now want to improve the random basis without calculating the SVD. To improve your basis you can use the following steps.

* Compute $\tilde{U}^{temp} = A\tilde{V}$.
* Compute a new orthogonal $\tilde{U}$ by orthogonalizing $\tilde{U}^{temp}$ using a **reduced** QR decomp
as $\tilde{U}, R = qr(\tilde{U}^{temp})$.
* Now compute $\tilde{V}^{temp} = A^T\tilde{U}$
* Obtain the new $\tilde{V}$ as $\tilde{V}, R = qr(\tilde{V}^{temp})$

To understand this procedure you can use an exact basis of singular vectors and write down on paper what happens. This algorithm is very similar to the power iteration for finding large eigenvalues of matrices. The reduced QR decomposition is obtained from the ``scipy.linalg.qr`` command by using the option ``mode="economic"``.

Repeat the above procedure a number of steps using a for-loop to improve your basis. How many basis vectors and loop iterations do you need to get an acceptable compressed image?



**Solution**

```python
                                                 '''Image Compression'''

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.misc

def proj_A_with_random_U_V(A,k):
    m,n = np.shape(A)
    U = scipy.linalg.qr(np.random.randn(m,k),mode = 'economic')[0]
    V = scipy.linalg.qr(np.random.randn(n,k),mode = 'economic')[0]
    proj_A = U@(U.T@A@V)@V.T
    plt.imshow(proj_A,cmap = 'gray')
    plt.title(r'Projected image A on random matrices U & V for $k$=%d' %(k))
    plt.axis('off')
    return plt.show()


def compression(A,k,trials):
    n = A.shape[1]
    V = scipy.linalg.qr(np.random.randn(n,k),mode = 'economic')[0]  #Pick a random nxk orthogonal matrix V
    for i in range(trials):
        U_temp = A@V 
        Q_U = scipy.linalg.qr(U_temp,mode='economic')[0]     #Orthogonalise U_temp
        V_temp = A.T@Q_U
        V = scipy.linalg.qr(V_temp,mode='economic')[0]    #Orthogonalise V_temp to form a new V
    Proj_A = Q_U@(Q_U.T@A@V)@V.T           #Projection of A onto spaces spanned of U,V
    plt.imshow(Proj_A,cmap='gray');
    plt.axis("off")
    return plt.show()

A = scipy.misc.face(gray=True)
print('Part 1: Here I used random matrices for U and V to project the image A on span of U and V')
print()
print(proj_A_with_random_U_V(A,1000))
print()
print('Using random matrices for U and V results in a very poor image quality even after k= 1000 chosen basis vectors')
print()
print('====================================================================================================================')
print()
print('Part 2')
print()
print()
[(plt.title(r'Projected Image A with $k$=%d basis vectors' %(k)),compression(A,k,2)) for k in [10,30,500]]
```
